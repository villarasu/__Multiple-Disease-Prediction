## feature & target

```python
X = df_parkinsons.drop(columns=['status'],axis=1)
Y = df_parkinsons['status']
```

## used scaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score,roc_auc_score
from imblearn.over_sampling import SMOTE
```

```python

# Split BEFORE scaling & SMOTE (to avoid data leakage)
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42, stratify=Y
)
```

```python
# Scale features
scaler = StandardScaler()
X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns)
X_test_scaled = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns)
```

```python
# Apply SMOTE only on training set
sm = SMOTE(random_state=42)
X_train_res, Y_train_res = sm.fit_resample(X_train_scaled, Y_train)
```

```python
# Train Random Forest
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train_res, Y_train_res)
```

```python
# Predictions
y_pred = clf.predict(X_test_scaled)
```

```python
# Evaluation
print("✅ Confusion Matrix:\n", confusion_matrix(Y_test, y_pred))
print("\n📊 Classification Report:\n", classification_report(Y_test, y_pred))
print("🎯 Accuracy:", accuracy_score(Y_test, y_pred))
print("🔵 ROC-AUC:", roc_auc_score(Y_test, clf.predict_proba(X_test_scaled)[:,1]))
```

## save model

```python
import pickle

# Save model
pickle.dump(clf, open("parkinsons_model.pkl", "wb"))

# Save scaler (important since you used StandardScaler)
pickle.dump(scaler, open("scaler.pkl", "wb"))

print("✅ Model and scaler saved successfully!")
```
